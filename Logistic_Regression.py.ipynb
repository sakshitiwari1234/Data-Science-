# Logistic Regression

introduction - > till here we completed the series of linear regression , and prpediction based on the random data which is continous in nature, whenever there is a need to build predictive models such as homeprice , stock market , weaterhf roecast etc, in which the data comes in numerical format underlying the limited set of output, we use of linear regression.

Now, when the prediction, is based on the categorical data , of only limited responses such as yes or no,
for example - spam mail prediction, will customer buy stock or not, the answer would be yes or no, only in such conditions we use , logistic regression().
will a peron can vote or not, what parties a person can vote etc etc.

such problems are called as classfication problem and are handled and solved using logistic regression techniques.


If the answer is Yes/No , it is called binary classification.
If the answer consist of more than two or three outputs , it is called as multiclass classification.


LogisticRegression is the technique to represent the data in the form of sigmoid curved made by sigmoid function..

performing the data analysis , predicting the output of whether the insurance is  bought by a certain age of people or not.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/logistic_regression.csv')
data
```

```python
data
```

```python
data.drop(['Unnamed: 3'], axis = 1, inplace = True)

```

```python
data.drop(['Unnamed: 4'], axis = 1, inplace = True)
```

```python
name = data.columns
name
data
```

```python
# to see the data ditribution
plt.scatter(data.age, data.insurance_b, marker = '+', color = 'r')
```

```python
# splitting our data
data.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data[['age']], data.insurance_b, train_size = 0.9)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
# now perfomring the training for our model.
model.fit(X_train, Y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, Y_test)
```

```python
model.predict_proba(X_test)
```
