```python
import numpy as np
import matplotlib.pyplot as plt

#setting width of the bar
bwidth = 0.25
fig = plt.subplots(figsize = (12,8))

it = [32,54,67,34,23]
cse= [67,43,32,23,43]
aiads = [67,46,23,67,44]

#setting posititon of bar in x axis
br1 = np.arange(len(it))
br2 = [x + barwidth for x in br1]
br3 = [x + barwidth for x in br2]



plt.bar(br1, it)
plt.bar(br2, cse)
plt.bar(br3, aiads)
plt.show()
```

**HISTOGRAM**
HISTOGRAM is the form of some viual represetnation of the data. when data is provided in some comon groups forms then hostogram is used to reprent that data visually. it is a type of plot where x-axis represents bin ranges and while y axis represents frequency..


x  -- bin ranges that is continous data
y  -- frequency distribution.


to create a histogram using matplotlib we use matplotlib.pyplot.hist()function used to create histogram of x.



Different methods if plotting histogram in matplolib are as follows

1. basic histogram.
2. customized histogram with density-plot.
3. customized histogram with watermark.
4. multiple hitogram with subplots.
5. stacked histogram.
6. 2d histogram.


question -- createa a basic hitogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# generating random data for histogram.
data = np.random.randn(1000)

#plotting a basic histogram
plt.hist(data, bins=30, color ='skyblue', edgecolor = 'black')

plt.xlabel('values')
plt.ylabel('frequency')
plt.title('basic histogram')

plt.show()

```

customized hitogram in matploplib with denity plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = np.random.randn(1000)

# creating customizd histogram with density plot
sns.histplot(data, bins = 30, kde = True, color = 'lightgreen', edgecolor = 'red'
)
#adding labels
plt.xlabel('values')
plt.ylabel('density')
plt.title('customized histogram with density plot')

plt.show()
```
