# Frequency **Table**

a frequency table is a table that dispplays the frequrny of different categories.this type of table is basically used for understanding the friequency distribution of values in a dataset.
let us see how to create a frequency distribution.

to find frequency of individual values in a panda serie , we use value_counts()function.

```python
import pandas as pd

#define series
data = pd.Series([2,4,5,6,2,4,3,4,5,6,7,5,4,3,2,4,5,6,6,5,4,5,3,5,5,4,6,3,2,1,6])

#findinf frequency of each values
data.value_counts(sort = False)
#sort= False is only used to show the numbers by default otherwsie it would show on the basis of number of frequency ditribution in ascedning order.
```

suppose we give a frequency for amle and female as 79 and 27 now make a frequency distrbution table for this now.



```python
import matplotlib.pyplot as plt

category = ['male', 'female']
frequency = [79, 27]

# plotting bar chart.
plt.bar(category, frequency)

plt.xlabel('gender')
plt.ylabel("frequency")
plt.title("frequency distribution plot for gender")

plt.show()
```

count unique words in a string
we are given some unique words in a string, we need to find out the most relevant words not used.

```python
import pandas as pd
def  unique_words(input_string):
  words = input_string.split()
  a = {}
  for word in words:
    if word not in a :
      a[word] = 1
    else:
        a[word]+= 1

  return a

input_string = 'hello my name is rahul rahul hello'
result  = unique_words(input_string)
print(result)
```

**To create a frequency distribution of continous quantitative variable such as age.**

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')

# creating an array wiht age
data = [21,24,24,26,31,32,20,24,32,36,42,24,43,45,18,52,55,53,41]


# create a histogram and setting the bins.
plt.hist(data, bins = 5 , color = 'hotpink', edgecolor = 'black')

plt.xlabel('Data Points')
plt.ylabel("Frequency")

plt.title("frequency distribution plot for AGE(continous qualitative data)")
```

```python
import seaborn as sns

# Sample data
data = [1, 1, 2, 2, 2, 3, 3, 4, 4, 5]

# Plotting the histogram
sns.histplot(data, edgecolor='blue')  # Change the edge color to blue
plt.title('Histogram with Custom Edge Color')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```
