### MATPLOTLIB LINE

matplotlib line this section is basically to understand the pratterns in which a graph line can be made.

```python
# to make the linestyle we use the keyword linestyle alias as 'ls'
import matplotlib.pyplot as plt
import numpy as np
x = np.array([34,5,5,66,54,33])
y = np.array([23,43,55,44,55, 33])
plt.plot(x , y, marker = 'o', ms = 11, mec = 'hotpink', mfc = 'red', ls = 'dotted')
plt.show()
```

# MATPLOTLIB SUBPLOT

BY USING MATPLOLIB LIBRARY, we can plot multtiple plots at a time , using ubplot() function.

with subplot function we can draw multiple plots in one figure.

```python
import matplotlib.pyplot as plt
import numpy as np

#plot1
x = np.array([2,3,2,5,6,7])
y = np.array([3,4,6,7,8,8])

plt.subplot(1,2,1)
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,2)
plt.plot(x,y)

#plot 2
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,1)
plt.plot(x,y)

plt.show()
```

subplot function-----

the subplt funcion takes three argument that descirbes the layout of the figure
1st argument - take rows.
2nd argument - take columns as reference between 2 plots.
3rd argument - take the index of the plot.

example- plt.subplot1,2,1

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2,4,6,7])
y = np.array([6,3,7,8])

plt.subplot(3,1,1)
plt.plot(x,y)
plt.title("first plot")
plt.grid()

#plot 2
x = np.array([3,5,6,7])
y = np.array([3,5,7,8])

plt.subplot(3,1,2)
plt.plot(x,y)

#plot 3

x = np.array([28,49,65,73])
y = np.array([6,3,7,8])

plt.subplot(3,1,3)
plt.plot(x,y)
plt.title("first plot")
plt.grid()

plt.title('another show of same graph')
plt.show()
```

IN PYTHON , we can use scatter()function in pyplot to create scatter plots.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2,4,5,6,7,8,5,4,2,2,3,4,5])
y = np.array([8,6,5,4,3,4,5,3,2,4,5,5,4])

plt.scatter(x,y, marker = 'd')
plt.show()
```
