### splitting the data into train and test data.

we need to train and test split the data. using whole model to predict the data is not a good practise. therefore we split the data into two parts - train and test data.

here we have the data of mmilage of cars , age and salary of data.

the milage , and age are the independent variable whereas , salary is the dependant variable.

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/data.csv')
```

```python
print(df.head(5))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

**car vs selling price**

```python
plt.scatter(df['milage'],df['sail price'])
```

**car age Vs Sell Price($)**

```python
plt.scatter(df['age'],df['sail price'])
```

```python
X = df[['milage', 'age']]
Y = df['sail price']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test , Y_train , Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
len(X_train)
X_train
```

```python
len(Y_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
```

```python
clf.fit(X_train,Y_train)
```

```python
clf.predict(X_test)
```

```python
Y_test
```

```python
clf.score(X_test, Y_test)
```

```python
clf.score(X_train,Y_train)
```
