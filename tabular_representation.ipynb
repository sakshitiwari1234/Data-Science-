Selecting desired Rows and columns in pandas dataframe using iloc.

iloc() function is used for the selection of data attributes based on position. it is similar to loc()function but it uses only integer values to make the selection.There are various methods to select the rows and columns an access them, they are-

1. select single row.
2. select multiple rows.
3. select multiple rows with all the columns.
4. select all the rows with some particular columnss.


```python
import pandas as pd
data = [('mina', 'cse', '32', 'jec'),('ria','aiads',23,'sati'),('tina','iot','22','sati'),('sakshi','aiads', 20,'sati'),('pratyush', 'aiml',20, 'gyan ganga')]
df = pd.DataFrame(data, columns = ['name','branch','age', 'college'])
# df = pd.DataFrame(data)
df
```

```python
#from this dataset i am selecting the partcular row third by accessing through 3 index in iloc function.

access = df.iloc[3]
access
```

```python
result = df.iloc[[2,4]]
result
```

```python
# selecting some multiple rows with some particular columns.
result = df.iloc[[2,3],[0,1]]
result
```

```python
result = df.iloc[:,[3]]
result
```
