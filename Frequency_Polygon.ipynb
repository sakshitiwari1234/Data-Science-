question 1 --> construct a frequency ploygon for the given data.

```python
import matplotlib.pyplot as plt
age = [4,7,15,25,28,36,55,45,88,75,43,56,89,98,55,54,65,77,88,81,82,89,92,98,65,76,73,72,84]
intervals  = [0,20,30,40,50,60,70,80,90,100]
plt.xticks(intervals)
y,edges,_= plt.hist(age, bins = intervals , histtype = 'step', edgecolor = 'k')
midpoints = 0.5*(edges[1:]+edges[:-1])
plt.plot(midpoints, y)
```
